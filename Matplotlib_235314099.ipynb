```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

data = pd.read_csv('dataPenjualanSetahun.csv')


print(data.to_string(index=False))
```

```python

plt.plot(data['bulan'], data['total_profit'])

plt.grid()
plt.title('Total Keuntungan setiap Bulan')
plt.xlabel('Bulan')
plt.ylabel('Total Keuntungan')
plt.xticks(range(1,13))
plt.show()
```

```python
plt.bar(data['bulan'], data['sabunmandi'])

plt.grid()
plt.title('Data Penjualan Sabun Mandi')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penjualan Sabun Mandi')
plt.xticks(range(1,13))
plt.show()
```

```python

jumlahTotal = data[['facecream', 'facewash', 'pastagigi', 'sabunmandi', 'shampoo', 'moisturizer']].sum()
uk=[0,0,0,0,0,0.2]
plt.pie(jumlahTotal, labels=jumlahTotal.index,explode= uk,autopct='%1.1f%%')
plt.title('Total produk')
plt.show()
```
